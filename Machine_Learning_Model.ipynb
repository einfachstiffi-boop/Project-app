```python
import mlcroissant as mlc
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python


# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/atharvasoundankar/global-music-streaming-trends-and-listener-insights/croissant/download')

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
record_set_df.head()



# Alle Spalten decoden, falls es Bytes sind
for col in record_set_df.columns:
    # Prüfen, ob die erste Zelle ein Byte ist
    if isinstance(record_set_df[col].iloc[0], bytes):
        record_set_df[col] = record_set_df[col].apply(lambda x: x.decode("utf-8"))

record_set_df=pd.DataFrame(record_set_df)

bins = {
    'Classical': 'Soft',
    'Jazz': 'Soft',
    'Country': 'Soft',
    'Pop': 'Modern',
    'EDM': 'Modern',
    'Reggae': 'Modern',
    'R&B': 'Modern',
    'Rock': 'Rock/Metal',
    'Metal': 'Rock/Metal',
    'Hip-Hop': 'Rock/Metal'
}
record_set_df["Global_Music_Streaming_Listener_Preferences.csv/Top+Genre"] = record_set_df["Global_Music_Streaming_Listener_Preferences.csv/Top+Genre"].map(bins)


# Define bins
bins = [0, 20, 40, 60, 120]

# Define names for each bin
labels = [
    "0-19", "20-39", "40-59",
    "60+"
]

# Apply pd.cut
record_set_df["Global_Music_Streaming_Listener_Preferences.csv/Age"] = pd.cut(record_set_df["Global_Music_Streaming_Listener_Preferences.csv/Age"], bins=bins, labels=labels, right=False)
record_set_df=pd.DataFrame(record_set_df)
record_set_df
```

```python

# Assuming 'record_set_df' is your original DataFrame from previous steps.
# The error 'NameError: name 'df' is not defined' indicates that 'df' was not created.
# We will use 'record_set_df' which was created in the previous cell.

columns_to_drop_multiple = [
    'Global_Music_Streaming_Listener_Preferences.csv/User_ID',
    'Global_Music_Streaming_Listener_Preferences.csv/Number+of+Songs+Liked',
    'Global_Music_Streaming_Listener_Preferences.csv/Discover+Weekly+Engagement+(%25)',
    'Global_Music_Streaming_Listener_Preferences.csv/Repeat+Song+Rate+(%25)',
    'Global_Music_Streaming_Listener_Preferences.csv/Subscription+Type',
    'Global_Music_Streaming_Listener_Preferences.csv/Listening+Time+(Morning%2FAfternoon%2FNight)'

]

df_multi_cleaned = record_set_df.drop(columns=columns_to_drop_multiple, axis=1)

print(f"Original DataFrame shape: {record_set_df.shape}")
print(f"DataFrame shape after dropping multiple column(s): {df_multi_cleaned.shape}")

print("Here are the first 5 rows of the DataFrame after deleting the specified columns:")
display(df_multi_cleaned.head())
```

```python
record_set_df = df_multi_cleaned.rename(columns={
    'Global_Music_Streaming_Listener_Preferences.csv/Age':'Age',
    'Global_Music_Streaming_Listener_Preferences.csv/Country':'Country',
    'Global_Music_Streaming_Listener_Preferences.csv/Streaming+Platform':'Streaming Platform',
    'Global_Music_Streaming_Listener_Preferences.csv/Top+Genre':'Target',
    'Global_Music_Streaming_Listener_Preferences.csv/Most+Played+Artist':'Favorite artist',
    'Global_Music_Streaming_Listener_Preferences.csv/Minutes+Streamed+Per+Day':'Minutes Streamed Per Day',
    'Global_Music_Streaming_Listener_Preferences.csv/Listening+Time+(Morning%2FAfternoon%2FNight)':'Listening Time',
})
```

```python
print(record_set_df.columns)
```

```python
distinct_counts = record_set_df.nunique()
unique_values = record_set_df["Target"].unique()
print(unique_values)

duplicate_counts = record_set_df.groupby(['Country', 'Streaming Platform','Favorite artist']).size().reset_index(name="count")
duplicate_counts
```

```python
y=record_set_df['Target']
x=record_set_df.drop(columns=['Target'],axis=1)
#Target label encoding
encoder = LabelEncoder()
y = encoder.fit_transform(y)

# Nur kategorische Spalten auswählen
cat_cols = ['Age','Country', 'Streaming Platform','Favorite artist']
# OneHotEncoder vorbereiten
encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")

# OneHot-Encoding anwenden
encoded = encoder.fit_transform(x[cat_cols])

# Neue Spaltennamen holen
encoded_cols = encoder.get_feature_names_out(cat_cols)
df_encoded = pd.DataFrame(encoded, columns=encoded_cols)

```

```python
y=pd.DataFrame(y,columns=['Target'])
y
```

```python
df_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_encoded, y, test_size=0.2, random_state=42, stratify=y
)
# Random Forest Modell trainieren

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))

```

```python

def error_matrix(y_true, y_pred):

    df = pd.DataFrame({"true": y_true, "pred": y_pred})
    errors = df[df.true != df.pred]

    return errors.value_counts().reset_index(name="count")


error_counts = error_matrix(y_test, y_pred)
print(error_counts)


```
